```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from random import random
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_to_load = pd.read_csv("data/mouse_drug_data.csv")
```

```python
clinical_trial_data_to_load =pd.read_csv("data/clinicaltrial_data.csv")
```

```python
# Combine the data into a single dataset

combined_data_df = pd.merge(clinical_trial_data_to_load, mouse_drug_data_to_load,  on="Mouse ID", how= "outer")
single_data_df = combined_data_df.loc[(combined_data_df["Tumor Volume (mm3)"]== 45.000000)]
single_data_df.head()

```

```python
#Tumor Response to Treatment

# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
stored_means = combined_data_df.groupby (['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)']
# Convert to DataFrame
stored_means_df = pd.DataFrame (stored_means)
# Preview DataFrame
stored_means_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stored_sems = combined_data_df.groupby (['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)']
# Convert to DataFrame
stored_sems = pd.DataFrame(stored_sems)
# Preview DataFrame
stored_sems.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
stored_means = stored_means.reset_index()
stored_means_tables = stored_means.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

stored_sems = stored_sems.reset_index()
stored_sems_tables = stored_sems.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked
stored_means_tables.head()
```

```python
print(stored_means_tables.iloc[-1])
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(stored_means_tables.index, stored_means_tables["Capomulin"], yerr=stored_sems_tables["Capomulin"], color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(stored_means_tables.index, stored_means_tables["Infubinol"], yerr=stored_sems_tables["Infubinol"], color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(stored_means_tables.index, stored_means_tables["Ketapril"], yerr=stored_sems_tables["Ketapril"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(stored_means_tables.index, stored_means_tables["Placebo"], yerr=stored_sems_tables["Placebo"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)
# Save the Figure
plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Days")
plt.grid(True)
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.savefig("data/chart1.png")
plt.show()
```

```python
#Metastatic Response to Treatment

# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_site_means = combined_data_df.groupby (['Drug', 'Timepoint']).mean()['Metastatic Sites']
# Convert to DataFrame
met_site_means_df = pd.DataFrame (met_site_means)
# Preview DataFrame
met_site_means_df.head()


```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_site_sems = combined_data_df.groupby (['Drug', 'Timepoint']).sem()['Metastatic Sites']
# Convert to DataFrame
met_site_sems = pd.DataFrame(met_site_sems)
# Preview DataFrame
met_site_sems.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_site_means = met_site_means.reset_index()
met_site_means_tables = met_site_means.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

met_site_sems = met_site_sems.reset_index()
met_site_sems_tables = met_site_sems.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

# Preview that Reformatting worked
met_site_means_tables.head()
```

```python
#Survival Rates

# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

count_mise = combined_data_df.groupby (['Drug', 'Timepoint']).count()['Mouse ID']
# Convert to DataFrame
count_mise = pd.DataFrame(count_mise)
# Preview DataFrame
count_mise.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_mise = count_mise.reset_index()
count_mise_tables = count_mise.pivot(index="Timepoint", columns="Drug")["Mouse ID"]

# Preview that Reformatting worked
count_mise_tables.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(count_mise_tables.index,100*count_mise_tables["Capomulin"]/25,color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.plot(count_mise_tables.index,100*count_mise_tables["Infubinol"]/25,color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.plot(count_mise_tables.index,100*count_mise_tables["Ketapril"]/25,color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.plot(count_mise_tables.index,100*count_mise_tables["Placebo"]/25,color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

# Save the Figure
plt.title("Survival During Treatment")
plt.ylabel("Survival Rate %")
plt.xlabel("Days")
plt.grid(True)
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.savefig("data/chart3.png")

# Show the Figure
plt.show()
```

```python
# Calculate the percent changes for each drug
tumor_percent_change =  100 * (stored_means_tables.iloc[-1] - stored_means_tables.iloc[0]) / stored_means_tables.iloc[0]
tumor_percent_change_sem =  100 * (stored_sems_tables.iloc[-1] - stored_sems_tables.iloc[0]) / stored_sems_tables.iloc[0]

# Display the data to confirm
tumor_percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
Percent_Changes= (tumor_percent_change["Capomulin"],
                 tumor_percent_change["Infubinol"],
                 tumor_percent_change["Ketapril"],
                 tumor_percent_change["Placebo"])
Percent_Changes
```

```python


# Splice the data between passing and failing drugs

#fig, ax = plt.subplots()
drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
x_axis = np.arange(len(drugs))


# Orient widths. Add labels, tick marks, etc. 
plt.title('Tumor Volume Change over 45 Day Treatment', fontdict = {'fontsize': 12, 'fontweight': 'bold'})

plt.ylabel('% Tumor Volume Change')
plt.xticks(np.arange(len(drugs)), drugs)
plt.grid(True)

# Use functions to label the percentages of changes


# Call functions to implement the function calls
# colors each bar appropriately based on + or - values by using a conditional

plt.bar(x_axis, Percent_Changes,color = ['red' if Percent_Changes[i] > 0 else 'green' for i in np.arange(len(drugs))])

count = 0

#loops through tumor change data and places the value in the correct position based on + or - values
for i in Percent_Changes:
    if i < 0:
        y_coor = -4
    else:
        y_coor = 2
    plt.text(count, y_coor, str(round(i, 1)) + '%', ha = 'center', color = 'white')
    count += 1
    
# Save the Figure
plt.savefig("data/chart4.png")


# Show the Figure

plt.show()
```
