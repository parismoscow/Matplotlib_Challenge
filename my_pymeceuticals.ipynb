```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_to_load = pd.read_csv("data/mouse_drug_data.csv")
```

```python
clinical_trial_data_to_load =pd.read_csv("data/clinicaltrial_data.csv")
```

```python
# Combine the data into a single dataset

combined_data_df = pd.merge(clinical_trial_data_to_load, mouse_drug_data_to_load,  on="Mouse ID", how= "outer")
single_data_df = combined_data_df.loc[(combined_data_df["Tumor Volume (mm3)"]== 45.000000)]
single_data_df.head()

```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
stored = combined_data_df.groupby (['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)']
stored.head()
# Convert to DataFrame
stored_df = pd.DataFrame (stored)
# Preview DataFrame
stored_df
```
